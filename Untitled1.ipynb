```python
hobbies=('swim','walk','sing','play','watch')
print(len(hobbies))
hobbies_list = list(hobbies)
hobbies_list.append('type')
hobbies = tuple(hobbies_list)
print(hobbies)
fhobbies=('laugh','text','sing','play')
print(fhobbies.similar(hobbies))
```

```python
hobbies_set = set(hobbies)
fhobbies_set = set(fhobbies)

similar_hobbies = hobbies_set.intersection(fhobbies_set)
different_hobbies = hobbies_set.difference(fhobbies_set)

print("Hobbies similar to fhobbies:", similar_hobbies)
print("Hobbies different from fhobbies:", different_hobbies)
```
